```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# 1. how many pages are declining vs growin
print(df["trend_direction"].value_counts(normalize=True))

active = df[df["impressions_90d"] > 100]
print(active["trend_direction"].value_counts(normalize=True))

# 3. avg position by trend direction
print(df.groupby("trend_direction")["avg_position"].mean())

print(df["content_age_days"].std())
```
